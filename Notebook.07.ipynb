```python
import os
import requests
import zipfile
import StringIO

NATIONAL_DATA_URL = 'https://www.ssa.gov/oact/babynames/names.zip'
NATIONAL_DATA_DIR = 'national_baby_names'

STATE_DATA_URL    = 'https://www.ssa.gov/oact/babynames/state/namesbystate.zip'
STATE_DATA_DIR    = 'state_baby_names'

def ls(path=os.curdir, sort=True): 
    files = os.listdir(path)
    if sort:
        files.sort()
    for f in files:
        print f

def head(path, n=10):
    fh = open(path, 'r') 
    count = 0
    line = None
    while count != n and line != '':
        count += 1
        line = fh.readline().rstrip()
        print line
    fh.close()
    
def tail(path, n=10):
    '''Display last n lines of file'''
    for line in open(path).readlines()[-n:]:
        print line.rstrip()
    
def download_and_extract_zipfile(url, destdir):
    print 'Downloading {} ...'.format(url)
    response = requests.get(url)
    zipdata  = zipfile.ZipFile(StringIO.StringIO(response.content))

    print 'Extracting {} ...'.format(url)
    zipdata.extractall(destdir)
```

```python
download_and_extract_zipfile(STATE_DATA_URL, STATE_DATA_DIR)
```

```python
import matplotlib.pyplot as plt
def count_occurrences_in_year(path, baby_name):
    fh = open(path)
    total = 0
    num = 0
    for f in fh.readlines():
        name, gender, amount = f.split(',')
        if name == baby_name:
            num = int(amount)
        total += int(amount)
    return num, total
            
```

```python
count_occurrences_in_year(os.path.join(NATIONAL_DATA_DIR, 'yob1880.txt'), 'Peter')
```

```python
def count_occurrences_in_years(path, years, baby_name):
    occurrences = []
    for year in years:
        occurrences.append(count_occurrences_in_year(os.path.join(path, 'yob{}.txt' .format(year)), '{}' .format(baby_name)))
    return occurrences
        
```

```python
count_occurrences_in_years(NATIONAL_DATA_DIR, range(1880, 1929), 'Peter')
```

```python
def plot_national_occurrences(data, years):
    occurrences = []
    for count in data:
        occurrences.append(count[0])
    plt.figure(figsize=(10, 8))
    plt.plot(years, occurrences, linewidth=4)
    plt.xlabel('Years', fontsize=24)
    plt.ylabel('Number of Occurrences', fontsize=24)
```

```python
years = range(1880, 2016)
data  = count_occurrences_in_years(NATIONAL_DATA_DIR, years, 'Kailee')
plot_national_occurrences(data, years)
```

```python
def plot_national_percentages(data, years):
    percentage = []
    for count in data:
        percentage.append(count[0] * 100.0 / count[1])
    plt.figure(figsize=(10, 8))
    plt.plot(years, percentage, linewidth=4)
    plt.xlabel('Years', fontsize=24)
    plt.ylabel('Percentage of Occurrences', fontsize=24)
```

```python
years = range(1880, 2016)
data  = count_occurrences_in_years(NATIONAL_DATA_DIR, years, 'Patrick')
plot_national_percentages(data, years)
```

```python
def count_yearly_occurrences_in_state(path, baby_name):
    fh = open(path)
    count = {}
    num = 0
    for f in fh.readlines():
        state, gender, year, name, occurrence = f.split(',')
        if name == baby_name:
            num = int(occurrence)
            year = int(year)
            count[year] = num
    return count
```

```python
count_yearly_occurrences_in_state('state_baby_names/CA.TXT', 'Peter')
```

```python
def count_yearly_totals_in_state(path): 
    fh = open(path)
    count = {}
    for f in fh.readlines():
        state, gender, year, name, occurrence = f.split(',')
        year = int(year)
        value = count.get(year, 0)
        value += int(occurrence)
        count[year] = value
    return count
```

```python
count_yearly_totals_in_state('state_baby_names/CA.TXT')
```

```python
def plot_state_occurrences(data):
    occurrences = data.values()
    years = data.keys()
    plt.figure(figsize=(10, 8))
    plt.plot(years, occurrences, linewidth=4)
    plt.xlabel('Years', fontsize=24)
    plt.ylabel('Number of Occurrences', fontsize=24)
```

```python
data = count_yearly_occurrences_in_state('state_baby_names/WA.TXT', 'Kailee')
plot_state_occurrences(data)
```

```python
def plot_state_percentages(data, totals):
    ''' Plot the percentages of occurrences for the state data '''
    d = sorted(data.items())
    t = sorted(totals.items())
    years = []
    occurrences = []
    totals = []
    percentage = []
    
    for key, value in d:
        years.append(key)
        occurrences.append(value)
        
    for key, value in t:
        totals.append(value)
        
    for occurrence, total in zip(occurrences, totals): #trying to use the items of two different lists in an equation
        percentage.append(occurrence * 100.0 / total)
    
    
    plt.figure(figsize=(10, 8))
    plt.plot(years, percentage, linewidth=4)
    plt.xlabel('Years', fontsize=24)
    plt.ylabel('Percentage of Occurrences', fontsize=24)
```

```python
data   = count_yearly_occurrences_in_state('state_baby_names/CA.TXT', 'Peter')
totals = count_yearly_totals_in_state('state_baby_names/CA.TXT')
plot_state_percentages(data, totals)
```
