```python
import tornado.web
import json


PORT = 3003
    
NUMBERRANGE = {
    "Seattle" : "40-51",
    "Chicago" : "29-39",
    "New York City" : "17-28"
}

class FormHandler(tornado.web.RequestHandler):
    def get(self):
        self.render('whichcity.html')

class ResultsHandler(tornado.web.RequestHandler):
    def get(self):
        total = 0
        for q in range(1, 18):
            q = "q{}" .format(q)
            a = self.get_argument(q)
            total = total + int(a)
        if total < 29:
            answer = "New York City"
            url = "https://static1.squarespace.com/static/55031eb7e4b0b4506cdb1e3d/t/5765992815d5db3461561f95/1466276153174/streets+of+manhattan"
        elif 28 < total < 40:
            answer = "Chicago"
            url = "https://static01.nyt.com/images/2015/08/30/travel/30LAKEMICHIGANJP6/30LAKEMICHIGANJP6-master675.jpg"
        else:
            answer = "Seattle"
            url = "http://www.travelaroundusa.com/uploads/2/4/0/0/24002860/297080914_orig.jpg"
        self.render('resultspage.html', total=total, answer=answer, url=url, numberrange=NUMBERRANGE)

        
Application = tornado.web.Application([
        (r"/", FormHandler),
        (r"/results", ResultsHandler),
        (r"/files/(.*)", tornado.web.StaticFileHandler,{'path':'.'})
],
        debug=True
                                     )


Application.listen(PORT)
```
